```python

plt.title('1-Bedroom Apartment Rent in Denver, Colordo 2012 - 2019')
plt.plot([1000, 1100, 1200, 1168, 1262, 1373, 1362])
plt.ylim([1000, 1400])
plt.xlim([0,7])
plt.ylabel('Rent ($)')
plt.xlabel('2012 - 2019')

plt.show()

print('0 = 2012, 1 = 2013, 2 = 2014, 3 = 2015, 4 = 2016, 5 = 2017, 6 = 2018, 7 = 2019 (projected)')
```

```python
#After studying the generally upward trends of migration after the passing of Amendment 64 in 2012, 
#It should come to no surprise that rent is on an incline, in direct correlation with the population climb,

#If we operate off the following basic economical assumptions:
#A) If there are more people coming into a new city, then they will be a higher demand for places to live
#B) If there is a higher demand for places to live, then there will be greater competition for residencies 
#C) If there is competition for residencies, then the prices will inflate 

#We can conclude based on the previous two graphs that the passing of Amendment 64 sparked both a migration 
#And a (potential) rent bubble within Denver, Colorado. 

#One crucial facet of Amendment 64 is the taxation on commercial marijuana. Let's say for a moment 
#no one at all moved to Colorado (or we had zero Net Migration), there would still be an increase in the
#amount of taxes generated by Sales of Marijuana by the Colorado natives and current residencies. This fact
#is independent of the population increase trends. That's an increase in money for the state of Colorado 
#By itself. 

#And, since there has in fact been an influx in the population of Colorado since Amendment 64, that's an 
#Even greater amount of revenue to be collected by the Colorado government to decide how to allocate the funds
#In the End. Let's take a look. 
```
