# Preferred Tempo in Top 100 Music 

I explored a dataset as provided by Swedish music streaming service Spotify's Open Source: a list of the Top 100 streamed songs in 2017, categorized by song name, artist, and numerical levels of danceability, energy, key, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo, duration, and time signatures. 

In this particular exploration, we'll look at tempos, the speed at which music is played, measured by beats per minute, the "pulse" of the song. We're looking for whether or not there are patterns in the Beats Per Minutes in the most popular global songs, keeping in mind that the Top 100 refers to songs most searched and listened to only (not complexity of composition or skill), but, for our case, it will be a decent gauge into any insights about preferred tempos in popular music, and compare their Danceability levels as well.

# 1)  What are the Most Occuring BPM's in the Top 100 Streamed Songs? 

```python
spt['tempo'].describe()
```

```python
x = spt['tempo']
plt.title('Tempo in Top 10 Songs from Spotify 2017')
plt.ylabel('Beats Per Minute')
plt.boxplot(x)
plt.show()
```

```python
spt.hist('tempo')
plt.xlabel('Beats Per Minute (BPM)')
plt.title('BPM in the Top 100 (Spotify 2017)')
plt.show()
```

About 50% of tempo rates for all 100 songs on the top chart hit 140 - 150 Beats Per Minute, with outliers hitting 75 BPM and 180 BPM, respectively. Let's check out those hits successful enough to end the year 2017 in the top 10% and see how they relate.
