```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
%matplotlib inline
spt = pd.read_csv('/Users/raychilds/Downloads/Spotify2k17.csv')
```

 #                                    Preferred Tempo in Top 100 Music 

I explored a dataset as provided by Swedish music streaming service Spotify's Open Data Source: a list of the Top 100 streamed songs by the end of 2017, categorized by ID, song name, artist and numerical levels of danceability, energy, key, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo, duration, and time signature. 

In this particular exploration, we'll look at tempo: the speed at which music is played, measured by beats per minute, the "pulse" of music. We're looking to see whether or not there are patterns in the beats per minute in the most streamed Western music, bearing in mind that the Top 100 refers to songs most searched and listened to, only, and not based on complexity, composition or skill. For our case it makes a decent gauge into any insights about preferred tempos in popular music, and then compare these to their Danceability (combination of tempo, beat stability, overall tempo) levels as well. 

# 1)  What are the Most Occuring Tempo's in the Top 100 Streamed Songs of 2017? 

```python
spt['tempo'].describe()
```

```python
(values, counts) = np.unique(spt['tempo'], return_counts=True)
ind = np.argmax(counts)
values[ind]
```

```python
x = spt['tempo']
plt.title('Tempo in Top 10 Songs from Spotify 2017')
plt.ylabel('Beats Per Minute')
plt.boxplot(x)
plt.show()
```

```python
spt.hist('tempo')
plt.xlabel('Beats Per Minute (BPM)')
plt.title('BPM in the Top 100 (Spotify 2017)')
plt.show()
```

About 50% of tempo rates for all 100 songs on the top chart hit 100 - 120 Beats Per Minute, with outliers hitting 80 BPM and 180 BPM, respectively. Let's check out how they hit in those songs successful enough to land the Top 10% at the end of 2017.

# 2) What Are the Tempos in the Top 10% of Spotify's 2017 Chart? 

```python
spt[0 : 10]['tempo'].plot()
plt.title('Tempo in the Top 10 Songs of 2017')
plt.xlim(0,9)
plt.ylabel('Beats Per Minute')
plt.xlabel('Top 10 Songs in 2017')
```

```python
spt[0:10]['tempo'].describe()
```

```python
x = spt[0:10]['tempo']
plt.title('Tempo in Top 10 Songs from Spotify 2017')
plt.ylabel('Beats Per Minute')
plt.boxplot(x)
plt.show()
```

```python
labels = '130 BPM', '140 BPM', '150 BPM', '160 BPM', '170 BPM', '180 BPM', '190 BPM'
sizes = [31, 20, 15, 9, 7, 4, 1]
explode = (0, 0, 0, 0, 0, 0, 0.5)      
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
print("Beats Per Minute in Top 100 Music (Spotify 2017)")
plt.show()
```

Again, we see that the majority of the tempos hit about 100 BPM, with outliers around 80 and 180.  

# 3) 'Danceability' is a Measure of Beat Strength, Tempo Stability, & Overall Tempo. In the Top 100 Songs, Where Are the Danceability Levels? 

```python
spt['danceability'].describe()
```

```python
(values, counts) = np.unique(spt['danceability'], return_counts=True)
ind = np.argmax(counts)
values[ind]
```

```python
spt['danceability'].plot()
plt.title('Danceability in Top 100 Streamed Songs')
plt.xlabel('Top 100 Songs')
plt.ylabel('Danceability Levels')
```

Danceability ranges are much more varied than tempo by itself. Part of this is accounted for the complexity of danceability (a combination of three musical attributes). But for the most part, danceability levels stay above .5, with a few outliers never dropping below .3. This isn't too surprising, as we can assume pop algorithms require at least some danceability, for the song to be a hit at public place, as in, music that gives people energy and doesn't use too many minor chords that generally make people more thoughtful than anything else. It doesn't seem like the best marketing tactic to play music in stores that increases thoughtfulness, as opposed to feeling confident and shopping-ready.
