# Heart Rates in Pop Music: An Analysis.

The biological reason for human being's affection toward music is a [hotly researched topic](http://www.pbs.org/wgbh/nova/blogs/secretlife/blogposts/the-evolution-of-music/). There is much work done on whether our biological rhythm in the autonomomic nervous system (heart rate) is correlated with musical tempos (both are measured in beats per minute (BPM). For this analysis, we will assume that it is, establish a baseline resting heart rate for the average person, and then see where the median and mode BPM's fall within Spotify's Top 100 Music Hit List from 2017. 

I've chosen a simple dataset of 130 patients resting heart rates from the American Heart Association ordered simply by anonymous patient number, and their results from a heart monitor, while no activity was being performed. It's important to note, these rates may or may not be slightly higher than rest, if for some reason the patient was nervous while in the doctor's office, something they may percieve as "unnatural", thus a higher heartbeat.

The second dataset is Spotify's Top 100 Songs from 2017, categorizing the music by acousticness, energy, beats per minute, etc. Spotify is currently the leading music streaming servince in the world, and providing a public API by artist, song, and charts.

# 1) What is the Average Resting Heart Rate in BPM? 

```python
hr.hist('HeartRate')
plt.xlabel('Heart Rates Per Minute')
plt.title('Average Heart Rates in 130 Patients')
plt.xlim(55,90)
plt.show()
```

```python
(values, counts) = np.unique(hr['HeartRate'], return_counts=True)
ind = np.argmax(counts)
values[ind]
```

```python
hr.describe()
```

The average heartrate of a person at rest is 73 heart beats per minute. Resting heartrates are typically measured using bell curves, as they are a naturally occurring biological phenomenon with few outliers.

# 2)  Is There a Preferred Tempo* in Top 100 Music? Where does this Number Lie in Relation to the Average Resting Heart Rate?

*Tempo (noun) : Speed at which music is played, measured in BPM

Let's search the entire top 100 list for songs with the same amount of beats per minute as the average resting person: 

```python
spt.hist('tempo')
plt.xlabel('Beats Per Minute (BPM)')
plt.title('BPM in the Top 100 (Spotify 2017)')
plt.show()
```

```python
labels = '130 BPM', '140 BPM', '150 BPM', '160 BPM', '170 BPM', '180 BPM', '190 BPM'
sizes = [31, 20, 15, 9, 7, 4, 1]
explode = (0, 0, 0, 0, 0, 0, 0.5)      
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
print("Beats Per Minute in Top 100 Music (Spotify 2017)")
plt.show()
```

```python
x = [75, 130, 140, 150, 160, 170, 180, 190]
y = [100, 31, 20, 15, 9, 7, 4, 1]
plt.ylabel('Beats Per Minute')
plt.xlim(10, 210)
plt.ylim(0,105)
plt.title('Beats Per Minute in Top 100 Music (Spotify 2017)')
plt.scatter(x,y)

plt.show()
```

# 3) What Are the BPM's in the Biggest Hits (Top 10% of Spotify's 2017 Chart) ? 

```python
spt[0 : 10]['tempo'].plot()
plt.title('Beats Per Minute in the Top 10 Songs of 2017')
plt.xlim(0,9)
plt.ylabel('Beats Per Minute')
plt.xlabel('Top 10 Songs in 2017')
```

```python
spt[0:10]['tempo'].describe()
```

```python
x = spt[0:10]['tempo']
plt.title('Tempo in Top 10 Songs from Spotify 2017')
plt.ylabel('Beats Per Minute')
plt.boxplot(x)
plt.show()
```

```python
(values, counts) = np.unique(spt[0:10]['tempo'], return_counts=True)
ind = np.argmax(counts)
values[ind]
```

### In the Top 10 of Spotify's Chart from 2017, the average tempo was 114 Beats Per Minute, 41 Beats above the average resting heart rate. However, the tempo most frequently found in the Top 10 was 80 Beats Per Minute, only 7 BPM's above a resting a heart rate. Also, remember that all 100 songs in the list were over the average person's resting heart rate. 

### So, if we do indeed enjoy music due to it's correlation with our biological rhythms, and if pop music is a decent meausre of the most listened to music due to it's deliberate commercialized nature, perhaps we do enjoy most tempos above our normal activity levels, music that gets our hearts pounding and circulations revved, music that makes us dance.
